```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Generating 1000 random numbers

```python
np.random.seed(0)
x = np.random.randn(1000)
y = np.random.randn(1000)
```

```python
plt.scatter(x,y)
```

# Gradient Descent Function

```python
def gradient_descent():
    slope = 0
    intercept = 0
    alpha = 0.01

    n = 1000
    cost = 0
    for i in range(n):
        y_pred = slope*x + intercept
        error = y - y_pred
        cost = (1/len(x)) * sum(error**2)
        a_diff = -(1/len(x)) * sum(error)
        b_diff = -(1/len(x)) * sum(error * x)
        intercept = intercept - alpha * a_diff
        slope = slope - alpha * b_diff

    return slope,intercept,cost
```

# Linear Regression Function

```python
def lin_reg():    
    m,c,cost = gradient_descent()
    y1 = m*x + c
    rmse = math.sqrt(cost)
    print(f"m={round(m,7)} c={round(c,8)} MSE={cost} RMSE={rmse}")
    plt.scatter(x, y)
    plt.plot(x, y1, "red")

lin_reg()
```

# Sklearn Implementation

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
x = x.reshape(-1,1)
y = y.reshape(-1,1)
reg.fit(x,y)
```

```python
reg.intercept_ # same as the intercept that was returned by the gradient descent function
```

```python
reg.coef_ # same as the slope that was returned by the gradient descent function
```
